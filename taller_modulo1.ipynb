### 1.1 — Listas y operaciones básicas

Tienes la siguiente lista con la esperanza de vida de 10 países (en años).

````python
# Lista dada — NO modificar
esperanzas_de_vida = [72.3, 81.5, 60.1, 78.9, 55.4, 83.2, 67.8, 74.0, 49.3, 79.6]
````

**Ejercicio 1.1** — Usando Python puro (sin Pandas ni funciones built-in como `sum()` o `max()`):
- Calcula el **promedio** de la lista usando un bucle `for`
- Encuentra el **valor máximo** usando un bucle `for`
- Encuentra el **valor mínimo** usando un bucle `for`

````python
%pip install pandas
````

````python
# Tu código aquí
import pandas as pd

esperanzas_de_vida = [72.3, 81.5, 60.1, 78.9, 55.4, 83.2, 67.8, 74.0, 49.3, 79.6]
suma = 0
contador = 0

for valor in esperanzas_de_vida:
    suma += valor
    contador += 1

# Promedio
promedio = suma / contador

# Máximo
maximo = esperanzas_de_vida[0]

for valor in esperanzas_de_vida:
    if valor > maximo:
        maximo = valor

# Mínimo
minimo = esperanzas_de_vida[0]

for valor in esperanzas_de_vida:
    if valor < minimo:
        minimo = valor

print(f"Promedio: {promedio:.2f}")
print(f"Valor Maximo: {maximo:.2f}")
print(f"Valor Minimo: {minimo:.2f}")
````

### 2.2 — Limpieza de datos

🔧 **Ejercicio 2.2** — El dataset tiene algunos valores nulos (¡datos reales siempre tienen este problema!).  
1. Rellena los valores nulos de columnas **numéricas** con la **media** de cada columna
2. Verifica que ya no queden nulos
3. Explica en una celda de texto: ¿Por qué usamos la media y no simplemente eliminamos esas filas?

````python
# Tu código aquí
# Cargar archivo

df = pd.read_csv("salud_mundial.csv")

# 1. Rellenar nulos en columnas numericas con la media 
colNumericas = df.select_dtypes(include=["number"]).columns

for columna in colNumericas:
    media = df[columna].mean()
    df[columna] = df[columna].fillna(media)

# 2. Verificar que ya no quedan nulos 
print("Valores nulos por Columna despues del Relleno: ")
print(df.isnull().sum())
````

### 2.3 — Estadísticas descriptivas

🔧 **Ejercicio 2.3**
1. Muestra el resumen estadístico completo del dataset
2. ¿Cuál es la esperanza de vida **promedio** mundial?
3. ¿Cuál es el país con **mayor** gasto en salud? ¿Y el de **menor**?
4. ¿Cuál es la **mediana** de la mortalidad infantil?

````python
# Tu código aquí
# Cargamos el dataset
df = pd.read_csv("salud_mundial.csv")

# Rellenamos nulos en columnas numericas con la media 
colNumericas = df.select_dtypes(include=["number"]). columns
for columna in colNumericas:
    media = df[columna].mean()
    df[columna] = df[columna].fillna(media)

# 1. Resumen estadistico completo del dataset
print("Resumen estadistico completo: ")
print(df.describe(include="all"))

# 2. Esperanza de vida promedio Mundial
promVida = df["esperanza_vida"].mean()
print(f"\nEsperanza de vida promedio mundial: {promVida:.2f}")

# 3. Pais con mayor y menor gasto de salud 
filaMaxGasto = df.loc[df["gasto_salud_usd"].idxmax()]
filaMinGasto = df.loc[df["gasto_salud_usd"].idxmin()]

print("\nPais con mayor gasto en salud: ")
print(filaMaxGasto["pais"], "-", filaMaxGasto["gasto_salud_usd"])

print("\nPais con menor gasto en salud: ")
print(filaMinGasto["pais"], "-", filaMinGasto["gasto_salud_usd"])

# 4. Mediana de la mortalidad infantil 
medianaMortalidadInfantil = df["mortalidad_infantil"].median()
print("\nMediana de la mortalidad Infantil: ")
print(medianaMortalidadInfantil)
````

### 2.4 — Filtrado y selección

🔧 **Ejercicio 2.4**

### 2.5 — Agrupaciones

🔧 **Ejercicio 2.5** — Usa `groupby()` para:
1. Calcular la **esperanza de vida promedio** por `region`
2. Calcular el **gasto en salud promedio** por `nivel_ingresos`
3. Calcular el **número de países** por región
4. ¿Qué región tiene la mayor esperanza de vida promedio? ¿Y la menor?

````python
# Tu código aquí
# Cargar archivo
df = pd.read_csv("salud_mundial.csv")

# Rellenar nulos en columnas numéricas con la media
columnas_numericas = df.select_dtypes(include=["number"]).columns
for columna in columnas_numericas:
    media = df[columna].mean()
    df[columna] = df[columna].fillna(media)

# 1. Esperanza de vida promedio por región
vidaPorRegion = df.groupby("region")["esperanza_vida"].mean()
print("Esperanza de vida promedio por región:")
print(vidaPorRegion)

# 2. Gasto en salud promedio por nivel de ingresos
gastoPorIngresos = df.groupby("nivel_ingresos")["gasto_salud_usd"].mean()
print("\nGasto en salud promedio por nivel de ingresos:")
print(gastoPorIngresos)

# 3. Número de países por región
paisesPorRegion = df.groupby("region")["pais"].count()
print("\nNúmero de países por región:")
print(paisesPorRegion)

# 4. Región con mayor y menor esperanza de vida promedio
regionMayorVida = vidaPorRegion.idxmax()
regionMenorVida = vidaPorRegion.idxmin()

print("\nRegión con mayor esperanza de vida promedio:")
print(regionMayorVida, "-", vidaPorRegion.max())

print("\nRegión con menor esperanza de vida promedio:")
print(regionMenorVida, "-", vidaPorRegion.min())
````
